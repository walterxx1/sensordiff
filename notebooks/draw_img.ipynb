```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
def plot_etth1_each(result_path, subject_key, num_idx):
    with h5py.File(result_path, 'r') as f_r:
        test_results = f_r['test_results']
        subject_data = test_results[subject_key]
        # print('check shape', subject_data['ori_slide'].shape, subject_data['future_pred'].shape)
        
        # context = subject_data['ori_slide'][num_idx,:168]
        # future = subject_data['ori_slide'][num_idx,168:]
        
        context = subject_data['ori_slide'][:200,:]
        future = subject_data['ori_slide'][-168:,:]
        future_pred = subject_data['future_pred'][num_idx,-168:]
        
        # context = subject_data['norm_slide'][:168,:]
        # future = subject_data['norm_slide'][168:,:]
        # future_pred = subject_data['norm_future_pred'][num_idx,:]
        

    seq_length, num_features = context.shape
    
    fig, axes = plt.subplots(3, 3, figsize=(15, 8))
    axes = axes.flatten()
    
    for i in range(num_features):
        axes[i].plot(range(seq_length), context[:,i], label='His', color='blue')
        # axes[i].plot(range(seq_length, seq_length*2), future[:,i], label='label', color='green')
        axes[i].plot(range(seq_length, seq_length+168), future[:,i], label='label', color='green')
        axes[i].plot(range(seq_length, seq_length+168), future_pred[:,i], label='pred', color='red', linestyle='--')
        
        axes[i].set_title(f'Feature {i+1}')
        axes[i].set_xlabel('Time Step')
        axes[i].set_ylabel('Value')
        axes[i].legend()
    
    plt.tight_layout()
    plt.suptitle(f'Comparison of Real and Predicted Ones')
    plt.show()
```

```python
result_path = './Experiments/etth1/etth1_ep181_result.h5'
subject_name = '1'
plot_etth1_each(result_path, subject_name, num_idx=0)
```

```python
result_path = './Experiments/etth1_trend/etth1_ep99_result.h5'
subject_name = '1'
plot_etth1_each(result_path, subject_name, num_idx=0)
```

# Comparation of nromal and trend

```python
def plot_etth1_each(result_path1, result_path2, subject_key, num_idx):
    with h5py.File(result_path1, 'r') as f_r:
        test_results = f_r['test_results']
        subject_data = test_results[subject_key]
        # print('check shape', subject_data['ori_slide'].shape, subject_data['future_pred'].shape)
        
        # context = subject_data['ori_slide'][num_idx,:168]
        # future = subject_data['ori_slide'][num_idx,168:]
        
        context = subject_data['ori_slide'][:201,:]
        future = subject_data['ori_slide'][200:,:]
        future_pred1 = subject_data['future_pred'][num_idx,-168:]
        
        # context = subject_data['norm_slide'][:168,:]
        # future = subject_data['norm_slide'][168:,:]
        # future_pred = subject_data['norm_future_pred'][num_idx,:]
        
    with h5py.File(result_path2, 'r') as f_r:
        test_results = f_r['test_results']
        subject_data = test_results[subject_key]
        
        future_pred2 = subject_data['future_pred'][num_idx,-168:]
        
    seq_length, num_features = context.shape
    
    fig, axes = plt.subplots(3, 3, figsize=(15, 8))
    axes = axes.flatten()
    
    for i in range(num_features):
        axes[i].plot(range(seq_length), context[:,i], label='His', color='blue')
        # axes[i].plot(range(seq_length, seq_length*2), future[:,i], label='label', color='green')
        axes[i].plot(range(seq_length, seq_length+168), future[:,i], label='label', color='green')
        axes[i].plot(range(seq_length, seq_length+168), future_pred1[:,i], label='ori', color='red', linestyle='-', linewidth=2.5)
        axes[i].plot(range(seq_length, seq_length+168), future_pred2[:,i], label='trend', color='black', linestyle='--', linewidth=2.5)
        
        axes[i].set_title(f'Feature {i+1}')
        axes[i].set_xlabel('Time Step')
        axes[i].set_ylabel('Value')
        axes[i].legend()
    
    plt.tight_layout()
    plt.suptitle(f'Comparison of Real and Predicted Ones')
    plt.show()
```

```python
result_path1 = './Experiments/etth1/etth1_ep181_result.h5'
result_path2 = './Experiments/etth1_trend/etth1_ep99_result.h5'
subject_name = '56'
plot_etth1_each(result_path1, result_path2, subject_name, num_idx=0)
```

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

def plot_comparisons_generate(result_path, subject_key, num_idx):
    with h5py.File(result_path, 'r') as f_r:
        test_results = f_r['test_results']
        subject_data = test_results[subject_key][:]
        
    final_img = subject_data[num_idx, -1, :]
    seq_length, channels = final_img.shape[0], final_img.shape[1]
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    
    for i in range(final_img.shape[-1]):
        axes[i].plot(range(seq_length), final_img[:,i], label='acc', color='blue')
        
        axes[i].set_title(f'Feature {i+1}')
        axes[i].set_xlabel('Time Step')
        axes[i].set_ylabel('Value')
        axes[i].legend()
    
    plt.tight_layout()
    plt.suptitle(f'Comparison of Real and Predicted Ones')
    plt.show()
```
